```python
# STEP 2: RESHAPE DATA FOR ML (FINAL)

import pandas as pd
import numpy as np

df = pd.read_csv("produce.csv")

# Clean columns
df.columns = df.columns.astype(str)

# Take first row (overall foodgrain production)
row = df.iloc[0]

# Keep only year columns (1993 onwards)
year_data = row[row.index.str.contains("-")]

final_df = pd.DataFrame({
    "year": year_data.index.str.replace("3-", ""),
    "production": year_data.values
})

final_df["year"] = final_df["year"].astype(int)
final_df["production"] = pd.to_numeric(final_df["production"], errors="coerce")

final_df = final_df.dropna()

final_df.head()
```

```python
# STEP 3: MACHINE LEARNING MODEL (FINAL)

from sklearn.linear_model import LinearRegression

X = final_df[['year']]
y = final_df['production']

model = LinearRegression()
model.fit(X, y)

# Predict future production
future_years = pd.DataFrame({'year': [2025, 2026, 2027]})
future_prediction = model.predict(future_years)

future_prediction
```
